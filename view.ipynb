```python
# subtract mean and divide by standard deviation
samples = [(s - np.mean(s)) / np.std(s) for s in samples]
```

```python
# resample to 6000 samples, and plot each one against original. Original is samples at 2 second intervals, adjust x axis to match
# do not plot on same figure
for i in range(len(samples)):
    resampled = signal.resample(samples[i], 6000)
    plt.plot(np.arange(0, 60, 60 / 6000), resampled, label=f"Resampled {i}")
    plt.legend()
    plt.show()
    plt.close()
```

```python
# save the resampled data to a single csv file
resampled = [signal.resample(s, 6000) for s in samples]

# add label column, all of them labelled as A
np.savetxt("resampled.csv", resampled, delimiter=",")
```
